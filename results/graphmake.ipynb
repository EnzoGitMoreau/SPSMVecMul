```python
import math
for file in [ 'standard.csv', 'newImpl.csv']:
    data_armpl = np.genfromtxt("../res/"+file, delimiter=",", usemask=True)
    clean_data_armpl = []
    for i in range(len(data_armpl.mask)):
        bool = data_armpl.mask[i]
        if not bool:
            clean_data_armpl.append(data_armpl.data[i])
    
    plt.plot(np.linspace(1,len(clean_data_armpl), len(clean_data_armpl)), clean_data_armpl, label=file)
plt.title('Time for 100 mul of a 12000x12000 matrix containing 900000 blocks with the number of threads')
plt.legend()

plt.show()
```
